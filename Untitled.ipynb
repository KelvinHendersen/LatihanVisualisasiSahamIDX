```python
#Refrensi -> https://realpython.com/beautiful-soup-web-scraper-python/
import requests
from bs4 import BeautifulSoup

URL = "https://www.idxchannel.com/stocks?index=BBCA&indexdetail=stockexchang3"
URL2 = "https://www.idxchannel.com/stocks?index=BBNI&indexdetail=stockexchang3"
URL3 = "https://www.idxchannel.com/stocks?index=BBRI&indexdetail=stockexchang3"
page = requests.get(URL)
page2 = requests.get(URL2)
page3 = requests.get(URL3)

soup = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(page2.content, "html.parser")
soup3 = BeautifulSoup(page3.content, "html.parser")

print(page.text)
```

```python
results = soup.find(id="right")
results2 = soup2.find(id="right")
results3 = soup3.find(id="right")
```

```python
print(results.prettify())
print(results2.prettify())
print(results3.prettify())
```

```python
saham_elements = results.find_all("div", class_="market-list--list")
saham_elements2 = results2.find_all("div", class_="market-list--list")
saham_elements3 = results3.find_all("div", class_="market-list--list")
```

```python
for saham_element in saham_elements:
    print(saham_element, end="\n"*2)

for saham_elements2 in saham_elements2:
    print(saham_elements2, end="\n"*2)

for saham_elements3 in saham_elements3:
    print(saham_elements3, end="\n"*2)
```
